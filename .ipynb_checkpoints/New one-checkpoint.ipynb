```python
!pip install openpyxl
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession, functions as F

EXPORT_DIR = "exports"
os.makedirs(EXPORT_DIR, exist_ok=True)
spark = SparkSession.builder.appName("StudentMentalHealthBigData").getOrCreate()
```

```python
data_path = "students_mental_health.csv"
df = spark.read.csv(data_path, header=True, inferSchema=True)
print("Columns:", df.columns)
df.show(5)
```

```python
missing_count = df.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in df.columns])
missing_count.show()

num_cols = ['Age', 'CGPA', 'Depression_Score', 'Anxiety_Score', 'Financial_Stress', 'Stress_Level', 'Semester_Credit_Load']
medians = {col: df.approxQuantile(col, [0.5], 0.01)[0] for col in num_cols}
df = df.fillna(medians)

cat_cols = [
    'Gender', 'Sleep_Quality', 'Physical_Activity', 'Diet_Quality',
    'Social_Support', 'Relationship_Status', 'Substance_Use',
    'Counseling_Service_Use', 'Family_History', 'Chronic_Illness',
    'Extracurricular_Involvement', 'Residence_Type', 'Course'
]
for c in cat_cols:
    df = df.fillna({c: "Unknown"})
```

```python
df_pd = df.toPandas()  # If not already done

# Calculate quantiles for Depression_Score
q33, median, q66 = df_pd["Depression_Score"].quantile([0.33, 0.5, 0.66]).values

# Create Depression_Level column
df_pd["Depression_Level"] = pd.cut(
    df_pd["Depression_Score"],
    bins=[-np.inf, q33, q66, np.inf],
    labels=["Low", "Moderate", "High"]
)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,3))
sns.histplot(df_pd['Depression_Score'], kde=True, bins=12)
plt.title("Depression Score Distribution")
plt.xlabel("Depression Score")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(5,3))
sns.countplot(x='Gender', data=df_pd)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(5,3))
sns.countplot(x="Depression_Level", data=df_pd, order=["Low","Moderate","High"])
plt.title("Depression Level Distribution")
plt.xlabel("Depression Level")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x="Depression_Level", y="CGPA", data=df_pd, order=["Low","Moderate","High"])
plt.title("CGPA by Depression Level")
plt.xlabel("Depression Level")
plt.ylabel("CGPA")
plt.show()

plt.figure(figsize=(8,5))
corr = df_pd[['Age', 'CGPA', 'Stress_Level', 'Depression_Score', 'Anxiety_Score', 'Financial_Stress', 'Semester_Credit_Load']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Key Features")
plt.show()

```

```python
# 1. Create Depression_Level (and/or Depression_Class) in your Spark DataFrame
q33, median, q66 = df.approxQuantile("Depression_Score", [0.33, 0.5, 0.66], 0.01)
df = df.withColumn(
    "Depression_Level",
    F.when(F.col("Depression_Score") <= q33, "Low")
     .when((F.col("Depression_Score") > q33) & (F.col("Depression_Score") <= q66), "Moderate")
     .otherwise("High")
)

# 2. Register (or re-register) the DataFrame as a SQL view
df.createOrReplaceTempView("students")

# 3. Now your SparkSQL queries will work!
spark.sql("SELECT Gender, COUNT(*) AS num_students FROM students GROUP BY Gender ORDER BY num_students DESC").show()
spark.sql("SELECT AVG(CGPA) as avg_cgpa, MIN(CGPA) as min_cgpa, MAX(CGPA) as max_cgpa FROM students").show()
spark.sql("SELECT Depression_Level, COUNT(*) AS count FROM students GROUP BY Depression_Level ORDER BY count DESC").show()
spark.sql("SELECT Gender, AVG(Stress_Level) as avg_stress, MIN(Stress_Level) as min_stress, MAX(Stress_Level) as max_stress FROM students GROUP BY Gender").show()
spark.sql("SELECT Course, AVG(Anxiety_Score) as avg_anxiety FROM students GROUP BY Course ORDER BY avg_anxiety DESC").show()
spark.sql("SELECT Residence_Type, AVG(Financial_Stress) as avg_fin_stress FROM students GROUP BY Residence_Type ORDER BY avg_fin_stress DESC").show()
spark.sql("SELECT Relationship_Status, COUNT(*) AS num_students FROM students GROUP BY Relationship_Status ORDER BY num_students DESC").show()
spark.sql("SELECT Gender, corr(CGPA, Anxiety_Score) as cgpa_anxiety_corr FROM students GROUP BY Gender").show()

```

```python
# 1. Depression Score by Gender and Residence Type
spark.sql("""
    SELECT Gender, Residence_Type,
           AVG(Depression_Score) as avg_dep_score,
           MIN(Depression_Score) as min_dep_score,
           MAX(Depression_Score) as max_dep_score,
           COUNT(*) as n_students
    FROM students
    GROUP BY Gender, Residence_Type
    ORDER BY Gender, Residence_Type
""").show()

# 2. Correlation between CGPA and Anxiety_Score by Gender
spark.sql("""
    SELECT Gender, corr(CGPA, Anxiety_Score) as cgpa_anxiety_corr
    FROM students
    GROUP BY Gender
""").show()

# 3. Percentage of Students in Each Depression Level by Course
spark.sql("""
    SELECT Course,
           Depression_Level,
           COUNT(*) AS n,
           ROUND(100.0 * COUNT(*) / SUM(COUNT(*)) OVER (PARTITION BY Course), 2) AS percent_within_course
    FROM students
    GROUP BY Course, Depression_Level
    ORDER BY Course, Depression_Level
""").show()

# 4. Top 5 Courses with Highest % of High Depression Level
spark.sql("""
    SELECT Course,
           ROUND(100.0 * SUM(CASE WHEN Depression_Level='High' THEN 1 ELSE 0 END) / COUNT(*), 2) as high_dep_percent
    FROM students
    GROUP BY Course
    ORDER BY high_dep_percent DESC
    LIMIT 5
""").show()

# 5. Stacked Bar Table: Social Support vs Depression Level
spark.sql("""
    SELECT Social_Support, Depression_Level, COUNT(*) as count
    FROM students
    GROUP BY Social_Support, Depression_Level
    ORDER BY Social_Support, Depression_Level
""").show()

# 6. Financial Stress by Relationship Status, ordered by avg stress
spark.sql("""
    SELECT Relationship_Status, AVG(Financial_Stress) as avg_fin_stress
    FROM students
    GROUP BY Relationship_Status
    ORDER BY avg_fin_stress DESC
""").show()

# 7. Subquery: Students above-average CGPA within each Depression Level
spark.sql("""
    SELECT s1.*
    FROM students s1
    JOIN (
        SELECT Depression_Level, AVG(CGPA) as avg_cgpa
        FROM students
        GROUP BY Depression_Level
    ) s2
    ON s1.Depression_Level = s2.Depression_Level
    WHERE s1.CGPA > s2.avg_cgpa
    ORDER BY s1.Depression_Level, s1.CGPA DESC
""").show(10)

# 8. Counseling Service Use by Depression Level (tabular for bar chart)
spark.sql("""
    SELECT Depression_Level, Counseling_Service_Use, COUNT(*) as count
    FROM students
    GROUP BY Depression_Level, Counseling_Service_Use
    ORDER BY Depression_Level, Counseling_Service_Use
""").show()

# 9. Pie Chart Table: Relationship Status as % of all students
spark.sql("""
    SELECT Relationship_Status,
           COUNT(*) AS n_students,
           ROUND(100.0 * COUNT(*) / (SELECT COUNT(*) FROM students), 2) AS percent
    FROM students
    GROUP BY Relationship_Status
    ORDER BY percent DESC
""").show()

# 10. Cluster Analysis: Mean CGPA, Stress, Depression Score per Cluster (if clustering done)
if 'prediction' in df.columns:
    spark.sql("""
        SELECT prediction as cluster,
               AVG(CGPA) as avg_cgpa,
               AVG(Stress_Level) as avg_stress,
               AVG(Depression_Score) as avg_dep
        FROM students
        GROUP BY prediction
        ORDER BY cluster
    """).show()
```

```python
quantiles = df.approxQuantile("Depression_Score", [0.33, 0.5, 0.66], 0.01)
q33, median, q66 = quantiles
print(f"Quantiles: 33%={q33}, median={median}, 66%={q66}")

df = df.withColumn("Depression_Class", (F.col("Depression_Score") > median).cast("int"))
df = df.withColumn(
    "Depression_Level",
    F.when(F.col("Depression_Score") <= q33, "Low")
     .when((F.col("Depression_Score") > q33) & (F.col("Depression_Score") <= q66), "Moderate")
     .otherwise("High")
)
df.groupBy("Depression_Class").count().show()
df.groupBy("Depression_Level").count().show()
```

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler, StandardScaler

categorical_cols = cat_cols
indexers = [StringIndexer(inputCol=c, outputCol=c+"_idx", handleInvalid='keep') for c in categorical_cols]
for indexer in indexers:
    df = indexer.fit(df).transform(df)
encoder = OneHotEncoder(
    inputCols=[c+"_idx" for c in categorical_cols],
    outputCols=[c+"_onehot" for c in categorical_cols]
)
df = encoder.fit(df).transform(df)

feature_cols = [
    'Age', 'CGPA', 'Stress_Level', 'Depression_Score', 'Anxiety_Score',
    'Financial_Stress', 'Semester_Credit_Load'
] + [c+"_idx" for c in categorical_cols]
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features_vec")
df = assembler.transform(df)
scaler = StandardScaler(inputCol="features_vec", outputCol="features", withMean=True, withStd=True)
scalerModel = scaler.fit(df)
df = scalerModel.transform(df)

onehot_features = [
    'Age', 'CGPA', 'Stress_Level', 'Depression_Score', 'Anxiety_Score',
    'Financial_Stress', 'Semester_Credit_Load'
] + [c+"_onehot" for c in categorical_cols]
assembler2 = VectorAssembler(inputCols=onehot_features, outputCol="features_onehot")
df = assembler2.transform(df)
```

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.ml.feature import PCA

kmeans = KMeans(featuresCol='features', k=3, seed=42)
kmeans_model = kmeans.fit(df)
df = kmeans_model.transform(df)

clustering_evaluator = ClusteringEvaluator(featuresCol='features', predictionCol='prediction', metricName='silhouette')
silhouette = clustering_evaluator.evaluate(df)
print("KMeans Silhouette Score:", silhouette)

pca = PCA(k=2, inputCol="features", outputCol="pca_features")
pca_model = pca.fit(df)
df = pca_model.transform(df)
pca_pd = df.select('pca_features','prediction').toPandas()
pca_pd['PCA1'] = pca_pd['pca_features'].apply(lambda x: x[0])
pca_pd['PCA2'] = pca_pd['pca_features'].apply(lambda x: x[1])
sns.scatterplot(data=pca_pd, x='PCA1', y='PCA2', hue='prediction', palette='tab10')
plt.title("KMeans Clusters (PCA 2D)")
plt.show()
```

```python
df.groupBy("Depression_Class").count().show()

train_df, test_df = df.randomSplit([0.8, 0.2], seed=42)
train_class_counts = train_df.groupBy("Depression_Class").count().toPandas()
if len(train_class_counts) < 2:
    print("Warning: Training set contains only one class. Classification skipped.")
else:
    for colname in ['prediction']:
        if colname in train_df.columns:
            train_df = train_df.drop(colname)
        if colname in test_df.columns:
            test_df = test_df.drop(colname)

    from pyspark.ml.classification import LogisticRegression, RandomForestClassifier, GBTClassifier

    log_reg = LogisticRegression(featuresCol="features", labelCol="Depression_Class", maxIter=20)
    log_model = log_reg.fit(train_df)
    preds_lr = log_model.transform(test_df)

    rf = RandomForestClassifier(featuresCol="features_onehot", labelCol="Depression_Class", numTrees=50)
    rf_model = rf.fit(train_df)
    preds_rf = rf_model.transform(test_df)

    gbt = GBTClassifier(featuresCol="features", labelCol="Depression_Class", maxIter=30)
    gbt_model = gbt.fit(train_df)
    preds_gbt = gbt_model.transform(test_df)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator

def print_classification_metrics(preds, model_name="", skip_roc_auc=False):
    if not skip_roc_auc:
        evaluator_roc = BinaryClassificationEvaluator(
            rawPredictionCol="rawPrediction", labelCol="Depression_Class", metricName="areaUnderROC"
        )
        try:
            roc_auc = evaluator_roc.evaluate(preds)
        except Exception:
            roc_auc = None
    else:
        roc_auc = None

    evaluator_acc = MulticlassClassificationEvaluator(labelCol="Depression_Class", predictionCol="prediction", metricName="accuracy")
    accuracy = evaluator_acc.evaluate(preds)
    precision = MulticlassClassificationEvaluator(labelCol="Depression_Class", predictionCol="prediction", metricName="precisionByLabel").evaluate(preds)
    recall = MulticlassClassificationEvaluator(labelCol="Depression_Class", predictionCol="prediction", metricName="recallByLabel").evaluate(preds)
    f1 = MulticlassClassificationEvaluator(labelCol="Depression_Class", predictionCol="prediction", metricName="f1").evaluate(preds)
    if roc_auc is not None:
        print(f"{model_name} -- ROC AUC: {roc_auc:.3f}, Accuracy: {accuracy:.3f}, Precision: {precision:.3f}, Recall: {recall:.3f}, F1: {f1:.3f}")
    else:
        print(f"{model_name} -- Accuracy: {accuracy:.3f}, Precision: {precision:.3f}, Recall: {recall:.3f}, F1: {f1:.3f} (ROC AUC not available)")
    return [model_name, accuracy, precision, recall, f1, roc_auc]

results = []
results.append(print_classification_metrics(preds_lr, "Logistic Regression"))
results.append(print_classification_metrics(preds_rf, "Random Forest"))
results.append(print_classification_metrics(preds_gbt, "Gradient Boosted Trees", skip_roc_auc=True))

comp_df = pd.DataFrame(results, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1', 'ROC_AUC'])
display(comp_df)

plt.figure(figsize=(7,4))
sns.barplot(data=comp_df.melt(id_vars=['Model'], value_vars=['Accuracy', 'F1']), x='variable', y='value', hue='Model')
plt.title("Model Comparison: Accuracy and F1")
plt.ylabel("Score")
plt.show()

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_true = preds_rf.select("Depression_Class").toPandas()
y_pred = preds_rf.select("prediction").toPandas()

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Low", "High"])
disp.plot(cmap="Blues")
plt.title("Confusion Matrix (Random Forest)")
plt.show()
```

```python
from pyspark.ml.regression import RandomForestRegressor, GBTRegressor, LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator

# Split (you can reuse splits, but this keeps things explicit)
train_r, test_r = df.randomSplit([0.8, 0.2], seed=100)

# --- Fix: Remove old "prediction" columns if they exist ---
for colname in ['prediction']:
    if colname in train_r.columns:
        train_r = train_r.drop(colname)
    if colname in test_r.columns:
        test_r = test_r.drop(colname)

results = []

# 1. Random Forest
rf_reg = RandomForestRegressor(featuresCol="features", labelCol="Depression_Score", numTrees=50)
rf_reg_model = rf_reg.fit(train_r)
preds_rf_reg = rf_reg_model.transform(test_r)
rf_rmse = RegressionEvaluator(labelCol="Depression_Score", predictionCol="prediction", metricName="rmse").evaluate(preds_rf_reg)
rf_r2 = RegressionEvaluator(labelCol="Depression_Score", predictionCol="prediction", metricName="r2").evaluate(preds_rf_reg)
results.append(["RandomForestRegressor", rf_rmse, rf_r2])

# --- Fix: Remove prediction col before fitting next model ---
if 'prediction' in test_r.columns:
    test_r = test_r.drop('prediction')

# 2. Gradient Boosted Trees
gbt_reg = GBTRegressor(featuresCol="features", labelCol="Depression_Score", maxIter=30)
gbt_reg_model = gbt_reg.fit(train_r)
preds_gbt_reg = gbt_reg_model.transform(test_r)
gbt_rmse = RegressionEvaluator(labelCol="Depression_Score", predictionCol="prediction", metricName="rmse").evaluate(preds_gbt_reg)
gbt_r2 = RegressionEvaluator(labelCol="Depression_Score", predictionCol="prediction", metricName="r2").evaluate(preds_gbt_reg)
results.append(["GBTRegressor", gbt_rmse, gbt_r2])

if 'prediction' in test_r.columns:
    test_r = test_r.drop('prediction')

# 3. Linear Regression
lr = LinearRegression(featuresCol="features", labelCol="Depression_Score", maxIter=50)
lr_model = lr.fit(train_r)
preds_lr_reg = lr_model.transform(test_r)
lr_rmse = RegressionEvaluator(labelCol="Depression_Score", predictionCol="prediction", metricName="rmse").evaluate(preds_lr_reg)
lr_r2 = RegressionEvaluator(labelCol="Depression_Score", predictionCol="prediction", metricName="r2").evaluate(preds_lr_reg)
results.append(["LinearRegression", lr_rmse, lr_r2])

# Results table
import pandas as pd
df_results = pd.DataFrame(results, columns=["Model", "RMSE", "R2"])
print(df_results)

```

```python
import pandas as pd

# --- Select columns for Tableau export ---
export_cols = [
    'Age', 'Gender', 'CGPA', 'Stress_Level', 'Depression_Score', 'Anxiety_Score',
    'Sleep_Quality', 'Physical_Activity', 'Diet_Quality', 'Social_Support',
    'Relationship_Status', 'Substance_Use', 'Counseling_Service_Use', 'Family_History',
    'Chronic_Illness', 'Extracurricular_Involvement', 'Residence_Type', 'Course',
    'Depression_Class', 'Depression_Level', 'Cluster_Label_KMeans',
    'Predicted_Risk_Class', 'Predicted_Depression_RF', 'Predicted_Depression_GBT', 'Predicted_Depression_LR'
]

# --- Convert to Pandas DataFrame ---
full_pd = df.select([c for c in export_cols if c in df.columns]).toPandas()

try:
    rf_reg_feature_importance = rf_reg_model.featureImportances.toArray()
    features_list = df.schema["features"].metadata["ml_attr"]["attrs"]["numeric"] + \
                    df.schema["features"].metadata["ml_attr"]["attrs"].get("binary", [])
    feat_names = [x['name'] for x in features_list]
    importance_df = pd.DataFrame({
        "Feature": feat_names,
        "RF_Reg_Importance": rf_reg_feature_importance
    }).sort_values(by="RF_Reg_Importance", ascending=False)
except Exception:
    importance_df = pd.DataFrame()

# --- Export to Excel: All data + feature importance ---
with pd.ExcelWriter("student_mental_health_allpreds_tableau.xlsx") as writer:
    full_pd.to_excel(writer, sheet_name="all_data", index=False)
    if not importance_df.empty:
        importance_df.to_excel(writer, sheet_name="feature_importance", index=False)

print("Exported: student_mental_health_allpreds_tableau.xlsx")
```
